```python
import numpy as np

def create_sequences(df, sequence_length):
    sequences = []
    labels = []
    grouped = df.groupby('subject_id')
    
    for patient_id, group in grouped:
        group = group.sort_values(by='enddateyear')
        label = group["opioid_abuse"].values  # Convert to NumPy array
        features = group.drop(columns=['subject_id', 'opioid_abuse']).values
        
        for i in range(len(features) - sequence_length):
            sequences.append(features[i:i + sequence_length])
            labels.append(label[i + sequence_length])  # Single label for each sequence

    return np.array(sequences), np.array(labels)

# Example usage
sequence_length = 10  # Example sequence length
x, y = create_sequences(df, sequence_length)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, SpatialDropout1D
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

# Load the dataset
df = pd.read_csv('your_dataset.csv')

# Group by patients and sort by time
df = df.sort_values(by=['subject_id', 'admittimeyear'])

# Normalize the data
scaler = MinMaxScaler()
normalized_df = scaler.fit_transform(df.drop(['subject_id', 'opioid_abuse'], axis=1))

# Add back the non-normalized columns
normalized_df = pd.DataFrame(normalized_df, columns=df.columns.drop(['subject_id', 'opioid_abuse']))
normalized_df['subject_id'] = df['subject_id'].values
normalized_df['opioid_abuse'] = df['opioid_abuse'].values

# Create sequences
sequence_length = 10
sequences = []
labels = []

for patient_id in normalized_df['subject_id'].unique():
    patient_data = normalized_df[normalized_df['subject_id'] == patient_id]
    for i in range(len(patient_data) - sequence_length):
        sequences.append(patient_data.iloc[i:i+sequence_length].drop(['subject_id', 'opioid_abuse'], axis=1).values)
        labels.append(patient_data.iloc[i+sequence_length]['opioid_abuse'])

sequences = np.array(sequences)
labels = np.array(labels)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(sequences, labels, test_size=0.2, random_state=42)
```
